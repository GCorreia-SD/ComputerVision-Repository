ED10 - TRANSFORMADA DE HOUGH

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import machinevisiontoolbox.base as mvb
from matplotlib import pyplot as plt
from machinevisiontoolbox import *

plt.close('all')
print('Resultados Obtidos no Programa: ')

### Leitura da imagem "church.png"
Church = Image.Read('church.jpg', dtype='float', gamma='sRGB')
print('Imagem Original: ')
Church.disp()
church_image = cv2.imread("church.jpg", cv2.IMREAD_GRAYSCALE)

### Aplicar o operador Canny para detecção de bordas
edges = cv2.Canny(church_image, 100, 200)

### Aplicar a Transformada de Hough
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=50, maxLineGap=10)
# Modificando os parâmetros threshol, minLineLenght e maxLineGap, altera-se a proporção das linhas vermelhas plotadas 

### Desenhar as linhas na imagem original
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(church_image, (x1, y1), (x2, y2), (0, 0, 255), 2)

### Exibindo as imagens do programa
# Exibindo a Imagem Original
plt.figure()
plt.imshow(church_image, cmap='gray')
plt.title('Imagem Original em Escala de Cinza')

# Exibindo a Imagem Binarizada
plt.figure()
plt.imshow(edges, cmap='gray')
plt.title('Imagem Original Binarizada (Canny)')

# Exibindo a Imagem com as Linhas Detectadas
plt.figure()
plt.imshow(church_image, cmap='gray')
for line in lines:
    x1, y1, x2, y2 = line[0]
    plt.plot([x1, x2], [y1, y2], 'r--', linewidth=2)
plt.title('Imagem Original com Linhas Detectadas')
plt.savefig('Church_com_detecção_de_linhas.png')

plt.show()
```
