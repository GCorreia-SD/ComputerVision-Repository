ED2 - HISTOGRAMA

```python
import numpy as np
from machinevisiontoolbox import *
import machinevisiontoolbox.base as mvb
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.close('all')
print('Resultados Obtidos no Programa: ')

# Leitura da imagem Castle.jpg
castle = Image.Read("castle.png")
castle.disp()

# Calculando o histograma pelo MachineVisionToolBox e armazenando na variável H (gráfico plotado em preto)
H = castle.hist()
H.plot('frequency', True, color='black')

# Analisando o Histograma, sabemos que o nível mais frequente de cinza ocorre no primeiro pico
pico = H.peaks()
M = pico[0]

# Pelo Histograma, podemos colocar o limite antes do início da subida do segundo pico gráfico
T = 210

# Calculando a Imagem Binária correspondente com o limite do Threshold
(castle >= T).disp()

# Plotando a Imagem Binária lógica:
plt.show()

# Informando os valores utilizados:
print('O Nível mais Frequente de Cinza ocorre em:', M)
print('Valor do Threshold = ', T)
```
