```python
import numpy as np
from machinevisiontoolbox import *
import machinevisiontoolbox.base as mvb
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.close('all')
print('Resultados Obtidos no Programa: ')

# Leitura da imagem Penguins.png
penguins = Image.Read("penguins.png")
penguins.disp()

# Obtendo os Kernels vertical e horizontal de Sobel
Ku = np.array([-1, 0, 1]) * (1/2) / 10
Kv = np.array([[-1], [0], [1]]) * (1/2) / 10

# Aplicando a função de convolução e armazenando em Iu e Iv
Iu = penguins.convolve(Ku)
Iv = penguins.convolve(Kv)

# Plotando e Salvando a imagem Horizontal
Iu_ivert = Iu * (-1)
Iu_ivert.disp(colormap='gray')
#plt.savefig('Iu.png')

# Plotando e Salvando a imagem Vertical
Iv_ivert = Iv * (-1)
Iv_ivert.disp(colormap='gray')
#plt.savefig('Iv.png')

# Plotando e Salvando a imagem com todas as Bordas
I = (Iu^2 + Iv^2).sqrt()
I.disp(colormap='gray')
#plt.savefig('I.png')


plt.show()
```
