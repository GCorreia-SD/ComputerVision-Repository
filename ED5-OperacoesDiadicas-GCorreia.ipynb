```python
import numpy as np
import matplotlib.pyplot as plt
from machinevisiontoolbox import *
import machinevisiontoolbox.base as mvb
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.close('all')
print('Resultados Obtidos no Programa: ')

### Abrindo o Video fornecido (Monocromática Flutuante)
traffic_sequence = VideoFile("traffic_sequence.mp4", mono=True, dtype="float")

#Contando o número total de quadros e Imprimindo a quantidade máxima contada
num_frames = sum(1 for _ in traffic_sequence)
print(f"Quantidade máxima de frames no vídeo: {num_frames}")

### Reabrindo o Video fornecido (Monocromática Flutuante) para retornar ao primeiro frame
traffic_sequence = VideoFile("traffic_sequence.mp4", mono=True, dtype="float")

### Gerando o Sequenciamento de Quadros para o Video Diferença entre Frames
subtractor = None
max_frames = 35  #Defindo o número máximo de quadros desejados para serem gerados
print(f"Quantidade escolhida de frames no vídeo: {max_frames}")


# Estruturando o Loop para as imagens
for frame_count, im in enumerate(traffic_sequence):
    #Configurando o primeiro quadro do video:
    if subtractor is None:
        subtractor = im

    #Configurando as imagens comparadas com o frame anterior
    else:
        diferenca_frames = im - subtractor
        T = 0.10 #Definindo o limite de Threshold
        traffic_threshold = (diferenca_frames >= T)
        traffic_threshold.disp(reuse=True, fps=traffic_sequence.fps)

    #Verificando se o número máximo de quadros foi atingido
    if frame_count + 1 > max_frames:
        break
```
