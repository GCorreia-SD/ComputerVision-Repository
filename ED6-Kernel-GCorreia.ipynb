```python
import numpy as np
import matplotlib.pyplot as plt
from machinevisiontoolbox import *
import machinevisiontoolbox.base as mvb
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.close('all')
print('Resultados Obtidos no Programa: ')

### Leitura da imagem flowers4.png
flowers4 = Image.Read("flowers4.png", mono=True, dtype="double")
print('Imagem Original das Flores')
flowers4.disp()

### Gerando um Kernel médio de tamanho 15x15 com volume unitário
kU = Image.Constant(15, 15, value=(1/(15**2)))
print('Kernel Médio de Volume Unitário (kU)')
kU.disp()
#plt.savefig('kU.png')

# Aplicando o Kernel kU na imagem original
imU = flowers4.convolve(kU)

### Gerando um filtro Gaussiano
kG = Kernel.Gauss(sigma=5, h=8)
print('Kernel Gaussiano (kG)')
idisp(kG)
#plt.savefig('kG.png')

# Aplicando o Kernel kG na imagem original
imG = flowers4.convolve(kG)

### Plotando as imagens imU e imG com os Kernels obtidos anteriormente
print('Imagem criada apartir do Kernel Médio de Volume Unitário (imU)')
imU.disp()
print('Imagem criada apartir do Kernel Gaussiano (imG)')
imG.disp()
#plt.savefig('imU.png')
#plt.savefig('imG.png')

print('Perceba que a imagem imU possui linhas e colunas embaçadas ao fundo que não existem na imG')
print('Isso ocorre porque a imagem imG utiliza um Kernel Gaussiano')

plt.show()
```
