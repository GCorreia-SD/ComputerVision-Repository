```python
import numpy as np
import matplotlib.pyplot as plt
from machinevisiontoolbox import *
import machinevisiontoolbox.base as mvb
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.close('all')
print('Resultados Obtidos no Programa: ')

### Leitura da imagem flowers9.png
flowers9 = Image.Read("flowers9.png")   #Lendo a imagem com o MachineVision Tool Box => histograma geral
flowers, _ = iread("flowers9.png")      #Lendo a imagem com o NumPy alterado => histogramas específicos
print('Imagem Original')
flowers9.disp()
#plt.savefig('original.png')


### Obtendo o Histograma Geral da Figura Original
H = flowers9.hist()
H.plot('frequency', True, color='black')
#plt.savefig('H.png')


### Esmiuçando a imagem para obtermos os parâmetros de imagem e histograma separadamente para o RGB
# Exibindo os planos de cada cor
vetor_RGB = np.array(["R", "G", "B"])

for i in range(flowers.shape[2]):
    plt.figure()
    
    # Obtendo o histograma para o plano de cor atual
    hist, bins, _ = plt.hist(flowers[:, :, i].flatten(), bins=256, range=[0, 256], color='black')
    
    plt.title(f"Histograma do Plano {vetor_RGB[i]}")  # R, G, B
    plt.xlabel("Valor do Pixel")
    plt.ylabel("Frequência")
    plt.grid(True)

    #Exibindo a imagem equivalente do respectivo plano RGB
    idisp(flowers[:, :, i])
    #plt.savefig(f'Plano {vetor_RGB[i]}.png')


plt.show()

```
